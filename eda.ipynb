```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('drugAddiction.csv', encoding='latin-1')
data.head()
```

```python
data.info()
```

```python
for col in data.columns:
    print(f"Unique values in '{col}' :")
    print(data[col].unique())
    print(' ' * 30)
```

```python
data = data.drop(columns=['Motive about drug', 'Mental/emotional problem'])
```

```python
data.info()
```

```python
data.columns
```

```python
mapping = {
    'Age': {'between 22 to 35 years': 0, '15 to 22 years': 1, 'between 35 to 48':2, 'above 48': 3},
    'Gender': {'Male': 0, 'Female': 1},
    'Education': {'Undergraduate': 0, 'H.S.C / A levels': 1, 'Postgraduate (MSc/PhD)': 2,
    'Primary Education': 3, 'S.S.C /O levels': 4},
    'Enjoyable with-': {'One drug': 0, 'I have no idea': 1, 'Multiple drugs': 2},
    'Live with': {'Hostel/Hall': 0, 'With Family/Relatives': 1},
    'Conflict with law': {'No': 0, 'Yes': 1},
    'Spend most time': {'Alone': 0, 'Friends': 1, 'Family/ Relatives': 2},
    'Failure in life': {'Yes': 0, 'No': 1},
    'Suicidal thoughts': {'Yes': 0, 'No': 1},
    'Family relationship': {'Satisfactory': 0, 'Average': 1, 'Communcaiton gap': 2},
    'Financials of family': {'Solvent': 0, 'Medium': 1, 'Rich / Strong': 2, 'Poor / weak': 3},
    'Addicted person in family': {'No': 0, 'Not sure': 1, 'Yes': 2},
    'no. of friends': {'Few (1-4)': 0, 'Many (more than 5)': 1},
    'Withdrawal symptoms': {'No': 0, 'Yes': 1, 'I have no idea': 2},
    'friends\u2019 houses at night': {"No, I don't": 0, 'Sometimes': 1, 'Yes, often I do.': 2},
    'Satisfied with workplace': {'No': 0, 'Yes': 1},
    'Case in court': {'No': 0, 'Yes': 1},
    'Living with drug user': {'Yes': 0, 'No': 1, 'Not sure': 2},
    'Smoking': {'Yes, every day.': 0, 'Yes, occasionally': 1, "No, I don't": 2},
    'Ever taken drug': {'Yes': 0, 'No': 1},
    'Friends influence': {'Yes, often they do': 0, "No, they don't": 1},
    'If chance given to taste drugs': {"I'll try": 0, "i don't know/ confused": 1, "no, i'll not": 2},
    'Easy to control use of drug': {"Yes, It's possible": 0, "No, It's not possible": 1},
    'Frequency of drug usage': {'Once/twice a week': 0, 'Occasionally': 1, 'Never/ Not applicable': 2, 'Regularly': 3}
}
```

```python
for column, column_mapping in mapping.items():
    data[column] = data[column].replace(column_mapping)
```

```python
plt.figure(figsize=(8, 4))
data['Ever taken drug'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.title('Ever taken drug')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
data['Suicidal thoughts'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.title('Suicidal thoughts')
plt.show()
```

```python
plt.Figure(figsize=(8, 4))
sns.countplot(data=data, x='Gender', palette='viridis')
plt.title('Gender')
plt.xlabel('Gender: 0 = male 1 = female')
plt.ylabel('count')
plt.show()
```

```python
plt.Figure(figsize=(8, 4))
sns.countplot(data=data, x='Financials of family', palette='viridis')
plt.title('Financials of family')
plt.xlabel('Gender: 0 = Solvent, 1 = Medium, 2 = Rich/Strong, 3 = Poor/weak')
plt.ylabel('count')
plt.show()
```

```python
plt.Figure(figsize=(8, 4))
sns.countplot(data=data, x='Living with drug user', hue='Ever taken drug', palette='viridis')
plt.title('Comparison')
plt.xlabel('0 = yes, 1 = no, 2 = not sure')
plt.ylabel('count')
plt.show()
```

```python
plt.Figure(figsize=(8, 4))
sns.countplot(data=data, x='Education', hue='Ever taken drug', palette='viridis')
plt.title('Comparison')
plt.ylabel('count')
plt.show()
```

```python
plt.Figure(figsize=(15, 14))
sns.heatmap(data.corr(numeric_only=True), annot=False, cmap='coolwarm')
plt.title('Correlation heatmap')
plt.show()
```
